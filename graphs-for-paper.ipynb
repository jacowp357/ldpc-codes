```python
%matplotlib notebook
from matplotlib.pyplot import *
from jupyterplot import ProgressPlot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.ticker import FormatStrFormatter
import scipy.stats as ss
import graphviz as gv
import scipy.stats as stats
import scipy.special as special
import itertools
import time, random
import math
import matplotlib.animation as animation
pd.options.display.max_columns = None
plt.rcParams.update({'font.size': 14})
plt.rcParams["font.family"] = "Times New Roman"
```

```python
mu = -2
mu1 = 2
variance = 3
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), color='black')
x = np.linspace(mu1 - 3*sigma, mu1 + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu1, sigma), color='black')
plt.axvline(x=2, linestyle='--', color='black', alpha=0.5)
plt.axvline(x=-2, linestyle='--', color='black', alpha=0.5)

ax = plt.gca()
ax.axes.get_yaxis().set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

labels = ['', '', '', r'-$\sqrt{E_b}$', r'$\tau$', r'$\sqrt{E_b}$', '']

ax.set_xticklabels(labels)

x_pos = 2.5
y_pos = 0.25
plt.text(x_pos, y_pos, r"$p(r|b_{1})$")
x_pos = -1.5
y_pos = 0.25
plt.text(x_pos, y_pos, r"$p(r|b_{0})$")
plt.savefig("foo.pdf", bbox_inches='tight')
plt.show()
```
